# **In this Notebook gonna practice the bivariate analysis**
**In bivariate analysis, we analyze two variables simultaneously to determine the empirical relationship between them. It can provide a clear understanding of the association between these two variables and how they interact with each other. This type of analysis is fundamental in statistical applications that deal with data interpretation and predictions **

# 1. Scatterplot: ( Numerical - Numerical) analysis

```python
sns.scatterplot(x="total_bill", y="tip", data=tips) #Numerical - Numerical bivariate analysis
```

**Multivariate Analaysis on the tips dataset below**

```python
#Now on the same code we will do do multivariate analysis
sns.scatterplot(x='total_bill', y='tip', hue= 'sex', style = 'smoker',size = 'size', data = tips)
```

#  **2. BarPlot ( Numerical - Categorical)**

```python
traindf.head(5)
```

```python
sns.barplot(x='pclass', y='fare',data=traindf) #this is the bivariate analysis
```

**Performign the multivariate analysis on the same dataset of traindf**

```python
sns.barplot(x='pclass',y='age',hue='sex', data=traindf)
```

# **3.Box Plot ( Numerical - Categorical )**

```python
sns.boxplot(x='sex', y ='age',data=traindf) # this is biavariate analysis
```

```python
sns.boxplot(x = 'sex', y='age', hue = 'survived',data = traindf) # this is bivariate analysis on boxplot
```

# **4. DistPlot ( Numerical - Categorical )**
The PDF is represented as a curve, and the probability of the variable falling within a certain range is given by the area under the curve within that range2. The total area under the curve is equal to 1, which corresponds to the total probability of all possible outcomes

```python
sns.distplot(traindf[traindf['survived']==0]['age']) #,hist=False
sns.distplot(traindf[traindf['survived']==1]['age']) #,hist=False
```

```python
sns.distplot(traindf[traindf['survived']==0]['age'],hist=False) #
sns.distplot(traindf[traindf['survived']==1]['age'],hist=False) #
```
