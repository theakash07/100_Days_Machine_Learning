## **FEATURE SCALING :- STANDARDISTAION**


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**CALLED A DATA VIA THE URL**

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/main/day24-standardization/Social_Network_Ads.csv"
headers = {"User-Agent":"Mozilla/5.0 (Machintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers = headers)
data = StringIO(req.text)
data2 = pd.read_csv(data)
```

```python
data2.head(5)
```

**REMOVED THE UNWANTED COLUMN**

```python
df = data2.iloc[:,2:]
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(df.drop('Purchased', axis = 1), df['Purchased'], test_size = 0.3, random_state = 0)
```

```python
X_train.shape
```

```python
X_test.shape
```

**Standard scaler Formula : Xi - X(mean)/ Standard deviation**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
# fit the scaler to the train set, it will learn the parameters
scaler.fit(X_train)
```

```python
# transform train and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
scaler.mean_
```

```python
X_train
```

```python
X_train_scaled
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```
