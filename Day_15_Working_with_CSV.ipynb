**Opening a Csv file with URL**

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/prm2711/python-recommender/master/zomato.csv"
headers = {"User-Agent":"Mozilla/5.0 (Machintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
request = requests.get(url, headers = headers)
data = StringIO(request.text)
data3 = pd.read_csv(data,encoding='latin-1')
```

```python
data3.head(5)
```

*Skipping Bad Lines*

```python
pd.read_csv('/content/placement.csv',error_bad_lines=False)
```

Convert datatypes into specific datatypes using *dtype* function

```python
pd.read_csv('/content/placement.csv',dtype={'placement':float})
```

Converter techniques

```python
def rename(name):
  if name == "Dark Green":
    return "DG"
  else:
    return name
```

```python
rename("Dark Green")
```

```python
pd.read_csv(data3,converters={'Rating color':rename})
```
