```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
from datetime import datetime

# Define project tasks and their timelines
tasks = [
    ("Research & Concept Finalization", "2025-02-01", "2025-02-15"),
    ("Initial Design & Prototyping", "2025-02-16", "2025-02-28"),
    ("Hardware & Software Integration", "2025-03-01", "2025-03-20"),
    ("Testing & Refinements", "2025-03-21", "2025-03-30"),
    ("Final Product & Documentation", "2025-03-31", "2025-04-06")
       ]

                    # Convert data into a DataFrame
df = pd.DataFrame(tasks, columns=["Task", "Start", "End"])

                    # Convert dates to datetime format
df["Start"] = pd.to_datetime(df["Start"])
df["End"] = pd.to_datetime(df["End"])
df["Duration"] = df["End"] - df["Start"]

                    # Create the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

                    # Plot each task as a bar
for i, (task, start, end) in enumerate(zip(df["Task"], df["Start"], df["End"])):
    ax.barh(task, (end - start).days, left=start, color="skyblue")

                        # Format the chart
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b %d"))
ax.xaxis.set_major_locator(mdates.DayLocator(interval=5))  # Adjust intervals for better visibility
plt.xlabel("Timeline")
plt.ylabel("Project Phases")
plt.title("Suraksha Project Timeline - Gantt Chart")
plt.grid(axis="x", linestyle="--", alpha=0.6)

                        # Show the Gantt chart
plt.show()
```
